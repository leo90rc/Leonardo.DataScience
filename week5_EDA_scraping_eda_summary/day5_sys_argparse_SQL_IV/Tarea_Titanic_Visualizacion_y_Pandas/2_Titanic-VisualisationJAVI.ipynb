2. Using matplotlib find the appropriate visualization to show the distribution of the column `age`

```python
#your code here
sns.set(style='ticks')
plt.hist(titanic['age'], bins= 8)
plt.xlabel('age')
plt.ylabel('number')
plt.title('age distribution')
plt.xlim(0)
```

3. Now use seaborn to show the distribution of the column `age`.

```python
#your code here
sns.set(style='ticks')
sns.distplot(titanic['age'], bins= 8, kde= False)
sns.despine()
```

4. Use subplots to plot the distribution of the Age variable with bins equal to 10, 20 and 50.

Use `sns.histplot()` for that aim. 

```python
#your code here
plt.figure(figsize = (8,10))
plt.subplot(3,1,1)
sns.histplot(titanic.age, bins= 10)
plt.xlim(0)

plt.subplot(3,1,2)
sns.histplot(titanic.age, bins= 20)
plt.xlim(0)

plt.subplot(3,1,3)
sns.histplot(titanic.age, bins= 50)
plt.xlim(0)
```

12. Using [`kdeplot`]("https://seaborn.pydata.org/generated/seaborn.kdeplot.html") from `seaborn` represent those who not survived distributed by age.

Hint: First you should "filter" the `titanic` dataset where the column "survived" is 0, indexing the column `"age"` only.

Arguments you should pass to the function: 

    - color = "red"
    - label = "Not Survived"
    - shade = True
    
You should get something like this: 

```python
survived = titanic[titanic['survived'] == 1]['age']
not_survived = titanic[titanic['survived'] == 0]['age']
```

```python
survived
```

![kdeplot](img/kdeplot.png)

```python
#your code 

sns.kdeplot(not_survived, color='red', fill=True, legend= True)
sns.kdeplot(survived, color='green', fill= True, legend= True)


```
