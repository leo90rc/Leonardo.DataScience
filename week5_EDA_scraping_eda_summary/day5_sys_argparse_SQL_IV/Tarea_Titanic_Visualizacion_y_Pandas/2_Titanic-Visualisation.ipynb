13. Using Seaborn plot the correlation matrix of various features.

Hint: search how to use `heatmap`



```python
titanic.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(titanic.corr(),
           vmin = -1,
           vmax = 1,
            annot = True,
           linewidths = .5)
```
