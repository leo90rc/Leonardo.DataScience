4. Use subplots to plot the distribution of the Age variable with bins equal to 10, 20 and 50.

Use `sns.histplot()` for that aim. 

```python
#your code here
plt.figure(figsize = (7,10))
plt.subplot(3,1,1)
sns.histplot(titanic.age, bins=10)
plt.xlim(0)

plt.subplot(3,1,2)
sns.histplot(titanic.age, bins=20)
plt.xlim(0)
plt.subplot(3,1,3)
sns.histplot(titanic.age, bins=50)
plt.xlim(0)
```

5. Whether the bin size affects your plot? Why?

El bin afecta al plot ya que, mientras mayor sea el valor que le ponemos, más subdivisiones tendremos, por lo cual veremos que la edad se dividirá. Esto lleva a que, se fraccione más el eje de las edades, y por lo tanto, se contabilicen menos edades para cada nuevo bin.

13. Using Seaborn plot the correlation matrix of various features.

Hint: search how to use `heatmap`



```python
titanic.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(titanic.corr(),
           vmin = -1,
           vmax = 1,
            annot = True,
           linewidths = .5)
```

14. According to the previous plot, what are the most correlated feature? Comment.

De acuerdo con el heatmap, podemos observar que las características más relacionadas son la clase con la tarifa. En este caso se puede apreciar una correlación negativa alta debido a la forma en la que se tienen en cuenta las clases. Es decir, la clase más adinerada es la que tiene asociado el número 1, y la menos adinerada la que tiene asociado el número 3. Entonces, en el plot podemos deducir que la clase "3" es la que cuenta con tarifas más bajas.

Otra relación que podemos encontrar es la de familiares a bordo, la cual tiene lógica ya que, si alguien viaja solo, sus variables "sibsp" y "parch" estarán altamente relacionadas, ya que ambas serán 0.
