### Step 1. Import the necessary libraries

```python

# remember to %matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

%matplotlib inline
```

### Step 2. Create the DataFrame it should look like below.

This time you will create the data.

```python
diccionario = {'first_name':['Jason', 'Molly', 'Tina', 'Jake', 'Amy'], 'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'], 'age': [42, 52, 36, 24, 73], 'female':[0, 1, 1, 0, 1], 'preTestScore': [4, 24, 31, 2, 3], 'postTestScore': [25, 94, 57, 62, 70]}
df = pd.DataFrame(diccionario)
df
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
step_3 = sns.FacetGrid(df)
step_3.map_dataframe(sns.scatterplot, x ='preTestScore', y ='postTestScore')
step_3.set_axis_labels('preTestScore', 'preTestScore')
step_3.
```

```python
step_3.map_dataframe(sns.relplot, x ='preTestScore', y ='postTestScore', s = df.age)
step_3.set_axis_labels('preTestScore', 'preTestScore')
```

```python
plt.scatter(df.preTestScore, df.postTestScore, s = df.age)
plt.title('preTestScore x postTestScore')
plt.xlabel('preTestScore')
plt.ylabel('preTestScore')
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
plt.scatter(df.preTestScore[df.female == 1], df.postTestScore[df.female == 1], s = df.postTestScore[df.female == 1] * 4.5)
plt.scatter(df.preTestScore[df.female == 0], df.postTestScore[df.female == 0], s = df.postTestScore[df.female == 0] * 4.5)
plt.title('preTestScore x postTestScore')
plt.xlabel('preTestScore')
plt.ylabel('preTestScore')
plt.fig
```

```python
plt.scatter(df.preTestScore, df.postTestScore, s= df.postTestScore * 4.5, c = df.female)
plt.title('preTestScore x postTestScore')
plt.xlabel('preTestScore')
plt.ylabel('preTestScore')
```
