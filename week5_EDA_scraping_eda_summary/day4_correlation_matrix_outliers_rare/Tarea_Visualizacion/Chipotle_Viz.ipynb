```python
top_ventas = chipo.groupby('item_name')['quantity'].sum().sort_values(ascending = False).iloc[0:5]

top_ventas = top_ventas.sort_values()
```

```python
plt.figure(figsize=(7,5))
plt.bar(top_ventas.index, top_ventas, width= 0.5)
plt.ylabel('Number of Times Ordererd')
plt.xlabel('Items')
plt.xticks(top_ventas.index, rotation = 90)
plt.title("Most Ordered Chipotle's Items")
```
