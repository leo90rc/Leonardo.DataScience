A partir del archivo food_100, se pide:

1. Limpiar el archivo borrando las columnas Unnamed.
2. ¿Qué porcentaje de valores NaN hay en cada columna?
3. ¿Tendría algún sentido clasificar el nombre de los alimentos a partir del top5 de columnas numéricas con menos valores NaN?

```python
import pandas as pd
import numpy as np
from pandas.api.types import is_numeric_dtype

df = pd.read_csv("../data/food_100.csv")
df
```

```python
for pos, value in enumerate(df.columns.str.contains('Unnamed')):
    if value == True:
        print('Posición columna Unnamed:', pos)
# Las siguientes son las columnas "Unnamed"
```

```python
df.drop(df.columns[0], axis=1, inplace= True)
df
```

```python
porcentaje_nan = {}
for i in range(df.shape[1]):
    porcentaje_nan[df.columns[i]] = int(df.isna().sum()[i] * 100 / df.shape[0])
```

```python
df_porcentajes = pd.DataFrame(list(porcentaje_nan.values()), index=porcentaje_nan.keys(), columns=['Porcentaje NaN (%)'])
df_porcentajes
```

```python
df_numeric_columns = df.select_dtypes('number')
df_numeric_columns
```

```python
porcentaje_nan_numeric_columns = {}
for i in range(df_numeric_columns.shape[1]):
    porcentaje_nan_numeric_columns[df_numeric_columns.columns[i]] = int(df_numeric_columns.isna().sum()[i] * 100 / df_numeric_columns.shape[0])
```

```python
lista_tuplas_numeric_columns = []
for i in porcentaje_nan_numeric_columns.items():
    lista_tuplas_numeric_columns.append(i)
lista_tuplas_numeric_columns
```

```python
len(lista_tuplas_numeric_columns)
```

```python
lista_tuplas_numeric_columns_sorted = sorted(lista_tuplas_numeric_columns, key = lambda x: x[1])
lista_tuplas_numeric_columns_sorted
```

```python
lista_tuplas_numeric_columns_sorted[0:6]
```

```python
for i in lista_tuplas_numeric_columns_sorted[0:6]:
    print('Columa:', i[0])
    print('Porcentaje NaN:', i[1], '%')
    print('-----------------------------------------------')
```
